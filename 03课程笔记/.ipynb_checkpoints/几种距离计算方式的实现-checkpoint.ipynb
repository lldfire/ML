### 1、闵可夫斯基距离：

#### 两个n维变量A(x11, x12,…,x1n)和B(x21,x22,…,x2n)闵可夫斯基距离的定义为：

    其中P是可变量：
    当p=1时，就是曼哈顿距离；
    当p=2时，就是欧氏距离；
    当p→∞时，就是切比雪夫距离；

```python
def minkowski_distance(vector_A, vector_B, p=2):
    """
    闵科夫斯基距离的计算方式，修改参数P可得到不同的距离结果，默认为欧式距离
    p = 1,曼哈顿距离；
    p = 2,欧式距离；
    p = no.inf,切比雪夫距离
    """
    assert len(vector_A) == len(vector_B), '两个矩阵长度不同，请检查！'
    vector_A = np.array(vector_A)   # 格式转换
    vector_B = np.array(vector_B)
    distance = np.linalg.norm((vector_A - vector_B), ord=p)
    return distance
```
