### 采样
#### 数据不均衡：数据分布不均匀，绝大多数数据都属于某个类别
1. 设置损失函数权重，使得少数类别的数据判断错误的损失大于多数数据判断错误的损失，从而导致模型向少数类别数据预测准确的方向偏
2. 下采样\欠采样：从多数类别的随机抽取样本，从而减少样本数量，使数据达到平衡
    - EasyEnsemble:采用不放回的数据抽取方式抽取多数样本，将其和少数样本组成一个模型；多次进行这样的操作从而构建多个模型；然后使用多个模型共同决策
    - BalanceCascade:利用Boosting增量思想训练模型：先通过下采样产生训练集，然后使用Adaboost算法训练一个分类器，然后使用该分类器对所有对大众数据进行预测，将预测正确样本从大众数据中剔除；重复上述两步操作，直到大众数据量等于小众数据量
3. ENN：对于多数类别样本数据而言，如果这个样本的大部分K近邻样本都和资深类别不一样，将其剔除，然后使用剔除后的样本训练模型
4. RENN：重复多次ENN操作，直到数据集无法被删除
5. 上采样：通过抽样的方式增加少数样本的数量，此法容易导致模型过拟合
6. 采用数据合成的方式生成更多的样本
7. 对于正负样本极不平衡的情况下，可以看成一分类或异常值检测问题，将少数样本当成异常数据

## 特征转换
1. 文本数据转换为数值类型的数据
    - 词袋法（BOW\TF）：统计词频, 忽略文本的语法和语序
    - TF-IDF：逆文档频率
    - HashTF: 不计算词频而是计算单词hash后的hash值的数量，运行速度快，但无法获取高频词
    - word2vec:
2. 缺省值填充
    - 重要、缺失率低
        - 计算填充
        - 通过经验或业务知识统计
    - 重要、缺失率高
        - 从其他渠道获取补全
        - 使用其他字段计算获取
        - 去除字段，并在结果中表明
    - 不重要、缺失率高
        - 去除该字段
    - 不重要、缺失率低
        - 不处理或简单填充
3. 定性属性哑编码
4. 定量特征属性二值化
    - 给定阈值，大于阈值设置为1，小于阈值设为0；1.对每个特征需要使用不同的阈值；2.一般情况下，不是进行二值化，而是分区化，即将一个连续数据，按照不同的区间，分为不同的级别
5. 特征标准化与归一化、正则化
    - 将矩阵的行转换为单位向量
