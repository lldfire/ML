# 第一章 统计学习方法概论
1. 统计学习是关于计算机基于数据构建概率统计模型并运用模型对数据进行分析与预测的一门学科。统计学习包括：监督学习、非监督学习、半监督学习、强化学习。
2. 统计学习的三要素：模型、策略、算法
    - 模型：所要学习的条件概率分布或决策函数，描述客观世界的数学模型，即潜在的数学规律，如朴素贝叶斯、隐马尔可夫等等
    - 策略：损失函数和风险函数
        - 0-1损失函数，平方损失函数，绝对随时函数，对数损失函数
        - 经验风险最小化和结构风险最小化
    - 算法：求解假设参数的方法，梯度下降法、EM算法、最大后验概率等
3. 最小二乘法拟合曲线

### 使用多项式拟合正弦函数

```python
# mu目标函数
def real_func(x):
    return np.sin(2 * np.pi * x)

# 多项式
def fit_func(p, x):
    f = np.poly1d(p)
    return f(x)

# 残差
def residuals_func(p, x, y):
    return fit_func(p, x) - y
    
```

```python
# 生成10个样本点
x = np.linspace(0, 1, 10)   # 等差数列，实例样本x
x_points = np.linspace(0, 1, 1000)
# 真实值
y_ = read_func(x)
# 加如正态分布噪音的目标函数的值
y = [np.random.normal(0, 0.1) + i for i in y_]

def fit(M=0):
    """M 为多项式的次数"""
    # 随机初始化多项式的参数
    p_init = np.random.rand(M + 1)
    # 最小二乘法,求解参数
    p_lsp = leastsq(residuals_func, p_init, args=(x, y))
    print('多项式拟合参数为', p_lsp)
    
    # 可视化
    plt.plot(x_points, real_func(x_points), label='real')
    plt.plot(x_points, fit_func(p_lsp[0], x_points), label='predict')
    plt.plot(x, y, 'bo', label='noise')
    plt.legend()
    return p_lsp
```

## 当 M过大时，会导致过拟合，需要使用正则化
- L1正则，L1范数：参数的绝对值，$\lambda$abs(p)
- L2正则，L2范数：参数的平方

```python
# 使用L2正则
regularization = 0.001
def residuals_func_regularization(p, x, y):
    ret = fit_func(p, x) - y
    ret = np.append(ret, np.sqrt(0.5 * regularization * np.square(p)))
    return ret
```

```python
# 最小二乘法加正则化
p_init = np.random.rand(9+1)
p_lsp_regularization = leastsq(
    residuals_func_regularization, p_init, args=(x, y))
```
