# 第二章 感知机学习策略
1. 感知机是根据输入实例的特征向量 x 对其进行二分类的线性分类模型：
$$f(x) = wx + b$$
分离超平面$wx + b = 0$
2. 感知机的学习策略是：极小化损失函数
3. 使用的算法是梯度下降法。
$$w = w + \eta y_i x_i$$
$$b = b + \eta y_i$$

## 原始形式（梯度下降）
### 例1:
正实例点：$x_1=(3, 3)^T, x_2=(4, 3)^T,负实例点：x_3 = (1, 1)^T$,使用原始形式求解感知机模型$f(x)=sign(wx + B)$.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['Heiti']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 加载数据集，仅选取两个分类的数据
iris = load_iris()
# data = pd.DataFrame(iris.data, columns=iris.feature_names)
# data['label'] = iris.target
# data.head(2)
x = iris.data
y = iris.target
print(iris.feature_names)
print(iris.target_names)
```

```python
# 数据集可视化
plt.scatter(x[:50, 0], x[:50, 1], label='setosa-0')
plt.scatter(x[50:100, 0], x[50:100, 1], label='versicolor-1')
plt.title('鸢尾花数据集')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
```

```python
# 数据集
x = np.array([x[:100, 0], x[:100, 1]]).T
y = np.array([1 if i == 1 else -1 for i in y[:100]])
```

### Perceptron

```python
class Perceptron_model:
    def __init__(self):
        self.w = np.ones(2, np.float32)
        self.b = 0
        self.learning_rate = 0.1
        
    def sign(self, x, w, b):
        y = np.dot(x, w) + b
        return y
    
    # 梯度下降法
    def fit(self, x_train, y_train):
        is_wrong = False
        while not is_wrong:
            wrong_count = 0
            for d in range(x_train.shape[0]):
                x = x_train[d]
                y = y_train[d]
                
                # 小于0 即，预测值和真实值异号，分类错误
                if y * self.sign(x, self.w, self.b) <= 0:
                    self.w = self.w + self.learning_rate * np.dot(y, x)
                    self.b = self.b + self.learning_rate * y
                    wrong_count += 1
            if wrong_count == 0:
                is_wrong = True
                
    def predict(self, x):
        if self.sign(x, self.w, self.b) > 0:
            return 1
        else:
            return -1
```

```python
perce = Perceptron_model()
perce.fit(x, y)
print('特征权重，', perce.w)
print('截距，', perce.b)

# 绘制分离平面
x_points = np.linspace(4, 7, 15)
y_ = -(perce.w[0] * x_points + perce.b) / perce.w[1]

plt.plot(x_points, y_)
plt.scatter(x[:50, 0], x[:50, 1], label='setosa-0')
plt.scatter(x[50:100, 0], x[50:100, 1], label='versicolor-1')
plt.title('鸢尾花数据集')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
plt.show()
```

### sklearn中的感知机

```python
from sklearn.linear_model import Perceptron
```

```python
clf = Perceptron(fit_intercept=True)
clf.fit(x, y)

print('特征权重', clf.coef_)
print('超平面截距', clf.intercept_)

# 画图
# 分离平面

x_points1 = np.arange(4, 8, 0.5)
y1_ = -(clf.coef_[0][0] * x_points1 + clf.intercept_[0])/clf.coef_[0][1]
plt.plot(x_points1, y1_)

plt.scatter(x[:50, 0], x[:50, 1], label='setosa-0')
plt.scatter(x[50:100, 0], x[50:100, 1], label='versicolor-1')
plt.title('鸢尾花数据集')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
plt.show()
```

```python
# 上图中有一个蓝点未正确分类，添加参数tol可修正
clf = Perceptron(fit_intercept=True, tol=None)
clf.fit(x, y)

print('特征权重', clf.coef_)
print('超平面截距', clf.intercept_)

# 画图
# 分离平面

x_points1 = np.arange(4, 8, 0.5)
y1_ = -(clf.coef_[0][0] * x_points1 + clf.intercept_[0])/clf.coef_[0][1]
plt.plot(x_points1, y1_)

plt.scatter(x[:50, 0], x[:50, 1], label='setosa-0')
plt.scatter(x[50:100, 0], x[50:100, 1], label='versicolor-1')
plt.title('鸢尾花数据集', fontsize=16, color='orange')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
plt.show()
```
