## 圆

```python
# 半径
r = 1
# 圆心
x0, y0 = 0, 0
num = [i for i in np.linspace(0, 2*np.pi, 10000)]
# print(num)
x = [x0 + r * np.cos(n) for n in num]
y = [y0 + r * np.sin(n) for n in num]
```

```python
# 画图
plt.plot(x, y, color='b')
plt.figure(num='fig1',figsize=(6,6),dpi=100,facecolor='#FFFFFF',edgecolor='#FF0000')
plt.show()
```

## 星型线

```python
a = 2
ttt = [i for i in np.arange(-10000, 10000, 0.1)]
x1 = [a * (np.cos(t) ** 3) for t in ttt]
y1 = [a * (np.sin(t) ** 3) for t in ttt]
```
