```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

from PIL import Image
from PIL import ImageFilter
```

### 图片预处理

```python
# 读取所有图片文件
image_path = '/Users/liuliangdong/project/jupyter_project/ML/项目实战/0.KNN识别验证码/result_images'
image_list = os.listdir(image_path)
```

```python
# 查看图片
test = Image.open(os.path.join(image_path, image_list[0]))

test_gray = test.convert('L')    # 灰度处理
plt.figure(facecolor='w')
plt.subplot(1, 2, 1)
plt.imshow(np.array(test))

plt.subplot(1, 2, 2)
plt.imshow(np.array(test_gray),cmap='gray')
plt.show()
```

```python
# 图片分割-找到分割点
plt.figure(facecolor='w')
split_lines = [5, 22, 31, 40, 44, 55]  
vlines = [plt.axvline(i, color='r') for i in split_lines] # 画出分割线
plt.imshow(np.array(test_gray), cmap='gray')
plt.show()
```
