```python
import numpy as np

from sklearn.datasets import load_boston
from matplotlib import pyplot as plt
from sklearn.svm import SVR    # svm的回归模型
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
```

```python
# 数据加载
X, Y = load_boston(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=9)
```

```python
# 使用管道进行交叉验证
models = [
#     Pipeline([
#         ('svr', SVR())
#     ]),
#     Pipeline([
#         ('ss', StandardScaler()),
#         ('svr', SVR())
#     ]),
    Pipeline([
        ('ss', StandardScaler()),
        ('pca', PCA()),
        ('svr', SVR())
    ])
]

# 参数，配置
parameters = {
    "svr__kernel": ['linear', 'rbf'],
    "svr__C": [0.1, 0.5, 0.8, 1, 2, 3, 4],
    "svr__gamma": [0.001, 0.01, 0.1, 1],
    "pca__n_components": [2, 4, 6, 7, 9, 10, 11, 13]
}
```

```python
for i in range(1):
    model = GridSearchCV(models[i], param_grid=parameters, cv=5)
    model.fit(x_train, y_train)
    print(f'第{i+1}个模型的最优参数为：', model.best_params_)
    print(f'第{i+1}个模型最优参数的评分是：', model.best_score_)    
```

```python
# 使用最优参数构建模型
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)

svm = SVR(C=4, gamma=0.1, kernel='rbf')
svm.fit(x_train, y_train)
```

```python
print(f'模型在预测集上的评分：{ svm.score(x_test, y_test)*100:.2f}%')
```

```python
y_predict = svm.predict(x_test)
```

```python
plt.rcParams['font.sans-serif'] = ['Heiti']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 画图
plt.figure(facecolor='w', figsize=(10, 6))
x_len = range(len(y_test))
plt.plot(x_len, y_predict, c='r', lw=2, label='预测值')
plt.plot(x_len, y_test, c='g', lw=2, label='真实值')
plt.legend()
plt.grid(axis='y')
plt.show()
```
