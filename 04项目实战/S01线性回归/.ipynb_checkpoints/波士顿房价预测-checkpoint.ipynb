```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import sklearn

# 导入线性回归，lasso回归(L1正则)， Ridge回归(L2正则)，弹性网络回归，
# CV表示使用交叉验证
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, ElasticNetCV
from sklearn.preprocessing import PolynomialFeatures  # 多项式扩展
from sklearn.pipeline import Pipeline      # 管道
from sklearn.model_selection import train_test_split  # 数据集划分
from sklearn.preprocessing import StandardScaler   # 数据标准化模块
from sklearn.model_selection import GridSearchCV   # 交叉验证模块
warnings.filterwarnings('ignore')   # 忽略警告
```

```python
def notempty(s):
    """过滤掉为空的函数"""
    return s != ''
```

```python
# 导入数据
df = pd.read_csv('./datas/boston_housing.data', header=None)
names = ['CRIM','ZN', 'INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']

# 创建一个指定维度的空数组,用来存放整理后的数据
data = np.empty((df.shape[0], 14))
for i, d in enumerate(df.values):
    d = map(float, filter(notempty, d[0].split(' ')))  # 去除空字符串
#     print(list(d))
    data[i] = list(d)
```

```python
# 分给数据 前13列为特征值，最后一列为目标值
x, y = np.split(data, (13,), axis=1) 
y.ravel()[0:5]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
# 创建管道
models = [
    Pipeline([
        ('ss', StandardScaler()),        # 数据标准化
        ('poly', PolynomialFeatures()),   # 多项式扩展
        ('linear', LinearRegression())     # 普通线性回归
    ]),
    Pipeline([
        ('ss', StandardScaler()),
        ('poly', PolynomialFeatures()),
        # Lass回归，其有个参数 alpha:float,将alpha设位0，则等同于普通线性回归
        ('linear', LassoCV(alphas=np.logspace(-3, 1, 10)))
    ]),
    Pipeline([
        ('ss', StandardScaler()),
        ('poly', PolynomialFeatures()),
        ('linear', RidgeCV(alphas=np.logspace(-3, 1, 10)))   # ridge(岭回归)
    ]),
    Pipeline([
        ('ss', StandardScaler()),
        ('poly', PolynomialFeatures()),
        # 弹性网络回归，两个参数子 alphas 和 l1_ratio： float, lasso 和ridge的权重比
        ('linear', ElasticNetCV(alphas=np.logspace(-3, 1, 10), l1_ratio=0.75))
    ])
]


# 参数设置
# parameters = {
#     "poly__degree": [3,2,1],    # 多项式展开阶数设置
#     "poly__interaction_only": [True, False],   # 空值是否产生类似 X1+X1的交互项
#     "Poly__include_bias": [True, False],   # 多项式幂为零的特征作为线性模型中的截距;true表示是，False表示否
#     "linear__fit_intercept": [True, False]
# }
parameters = {
    "poly__degree": [3, 2, 1],
    "poly__interaction_only": [True, False],  # 不产生交互项，如X1*X1
    "poly__include_bias": [True, False],  # 多项式幂为零的特征作为线性模型中的截距;true表示包含
    "linear__fit_intercept": [True, False]
}
```

```python
plt.rcParams['font.sans-serif'] = ['Heiti']  
plt.rcParams['axes.unicode_minus'] = False
```

```python
titles = ['Regression', 'Lasso', 'Ridge', 'ElasticNetCV']
colors = ['k', 'b', 'g', 'm']
plt.figure(figsize=(12, 8), facecolor='w')

# 绘制真实值的图
len_x = range(len(X_test))
plt.plot(len_x, Y_test, 'r', lw=2, label='真实值')
for i in range(4):
    # 获取模型并设置参数
    # 第一个参数，选择模型
    # param_grid：设置模型的参数，主要是多项式展开阶数
    # cv: 进行几折交叉验证
    model = GridSearchCV(models[i], param_grid=parameters, cv=5, n_jobs=1)
    # 模型训练
    model.fit(X_train, Y_train)
    print('%s 模型的最优参数是' % titles[i], model.best_params_)
    print("%s算法:R值=%.3f" % (titles[i], model.best_score_))
    
    # 模型预测
    y_predict = model.predict(X_test)
    plt.plot(len_x, y_predict, c=colors[i], lw=2, label=titles[i])
    
plt.legend(loc='uper left')
plt.grid(axis='y')
plt.title('波士顿房价预测模型预测结果展示')
plt.show()
```
