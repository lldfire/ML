欠拟合：模型训练集表现不好，那么测试集效果也不会好；
过拟合：模型训练集表现很好，测试集效果不好，
如何解决：1.增加样本数量；2.增加特征数量（数据维度）；
    在没有新的样本时，只能增加数据维度，多项式扩展可以达到增加数据维度的要求
详细介绍：https://www.jianshu.com/p/bad47166a4c0

为了解决过拟合：我们可以在损失函数中加入惩罚项，对于系数过大的二惩罚，主要有L1正则(L1-norm) 和 L2正则(L2-norm):
    l1-norm 为系数的 绝对值和，L2-norm 为系数的平方和
    L2-norm 被称岭回归（ridge)  L1-(lasso)

```python
# 导入相关包
import sklearn
from sklearn.model_selection import train_test_split   # 训练集测试集划分
from sklearn.linear_model import LinearRegression      # 线性回归模型
from sklearn.preprocessing import StandardScaler       # 数据标准化
from sklearn.pipeline import Pipeline                  # 管道
from sklearn.preprocessing import PolynomialFeatures
from matplotlib import pyplot as plt

import pandas as pd
import numpy as np
import time
```

```python
# 显示中文设置
plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 定义时间格式化输出的函数
def date_format(ser):
    """接收一个series对象，将时间格式化
    return：tuple
    """
    t = time.strptime(' '.join(ser), '%d/%m/%Y %H:%M:%S')
    return (t.tm_year, t.tm_mon, t.tm_mday, t.tm_hour, t.tm_min, t.tm_sec)
```

```python
# 读取数据集
path = r'./datas/household_power_consumption_1000.txt'
df = pd.read_csv(path, sep=';', low_memory=False, usecols=[0, 1, 2, 3, 4, 5])   # 仅读取前6列数据
names = df.columns.tolist()    # 列名
print(names)

# 处理异常数据
df.replace('?', np.nan, inplace=True)    # 将非法字符替换为nan
df.dropna(axis=0, how='any', inplace=True)             # 删除数据中包含nan的数据 

# 取出特征值和目标值
X = df[names[:2]]
X = X.apply(lambda x: pd.Series(date_format(x)), axis=1)   # 按行进行
Y = df[names[4]]
```

```python
# 设置1-5阶的多项式扩展模型，选择最优模型
# 划分训练集和测试集
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=1)

# 数据标准化，也可将 标准化过程放在管道中进行
# ss = StandardScaler()    # 对象的实例化
# X_train = ss.fit_transform(X_train)    # 训练集训练并转换
# X_test = ss.transform(X_test)    # 在模型构建数据的基础上 标准化测试集

# 设置管道 时间和电压的关系（linear-多项式）
models = [
    Pipeline([
        ('ss', StandardScaler()),    # 第一个操作，数据标准化
        ('Poly', PolynomialFeatures(degree=2)),    # 第二个操作，多项式扩展
        ('Linear', LinearRegression(fit_intercept=False))  # 最后一个操作，线性回归模型
    ])
]

model = models[0]
# 模型的训练
t = np.arange(len(X_test))   #
# 设置1-5阶多项式扩展，
N = 5
d_pool = np.arange(1, N, 1)  # 阶数设置

m = d_pool.size
colors = []  # 设置线条颜色
for c in np.linspace(16711680, 255, m):    # 生成一组等差数列
    colors.append('#%06x' % int(c))        # 将整数转化成16进制颜色
line_width = 3     # 线宽

plt.figure(figsize=(12, 6), facecolor='w')
for i, d in enumerate(d_pool):
    # 绘制真实值的曲线图
    plt.subplot(N-1, 1, i+1)    # 三个参数，前2个分别是图像的行数和列数， 第三个是本次绘图的位置
    plt.plot(t, Y_test, 'r-', label='真实值', ms=10, zorder=N)

    # 对Poly模型的参数赋值  模型名__参数值
    model.set_params(Poly__degree=d)     # 展开多项式的阶乘
    # 将训练集的特征值和目标值进行训练
    model.fit(X_train, Y_train)

    # model.get_params()  调用管道模型中的所有参数
    lin = model.get_params()['Linear']
    lin = model.get_params()['Linear']    # 调取linear参数
    output = u'%d阶，系数为：' % d
    # 判断Linear模型中是否有此alpha参数
    if hasattr(lin, 'alpha_'):
        idx = output.find(u'系数')    # 找到 ‘系数’ 的索引
        output = output[:idx] + (u'alpha=%.6f, ' % lin.alpha_) + output[idx:]
    # 判断Linear模型中是否有l1_ratio这个参数
    if hasattr(lin, 'l1_ratio_'):
        idx = output.find(u'系数')
        output = output[:idx] + (u'l1_ratio=%.6f, ' %
                                 lin.l1_ratio_) + output[idx:]

    print(output, lin.coef_.ravel())

    # 进行结果预测 和 评分
    y_hat = model.predict(X_test)
    score = model.score(X_test, Y_test)

    z = N - 1 if (d == 2) else 0
    label = u'%d阶, 准确率=%.3f' % (d, score)
    plt.plot(t, y_hat, color=colors[i], lw=line_width,
             alpha=0.75, label=label, zorder=z)
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.ylabel(u'%d阶结果' % d, fontsize=12)

# 预测值和实际值画图比较
plt.suptitle(u"线性回归预测时间和功率之间的多项式关系", fontsize=20)
plt.grid(b=True)
plt.show()
```
