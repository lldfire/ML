```python
import numpy as np

from sklearn.naive_bayes import GaussianNB   # 高斯朴素贝叶斯
from sklearn.naive_bayes import MultinomialNB    # 多项式朴素贝叶斯
from sklearn.naive_bayes import BernoulliNB    # 伯努利朴素贝叶斯
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

### GaussiaanNB
#### 简介
高斯朴素贝叶斯，可在线更新模型，适合处理连续数据

#### 参数
- priors: 类的先验概率，如果指定，则不会根据数据进行调整
- var_smoothing: 添加到所有特征最大方差部分，默认1e-9

#### 属性
- class_prior_: 每一个类的概率
- class_count_: 训练样本中每个类的观测数量
- theta_: 每个类中每个特征的平均值
- sigma_: 每个类中每个特征的方差
- epsilon_: 方差的绝对值的累加

#### 方法
- partial_fit(): 新增一批样本训练拟合模型，模型可进行在线学习，数据过大时也可分批簇训练减少模型训练的开销
- predict_proba(): 返回预测结果的概率值

```python
# 读取数据集
iris = load_iris()
X = iris.data
Y = iris.target
```

```python
# 数据预处理
X = StandardScaler().fit_transform(X)

x_train, x_test, y_train, y_test = train_test_split(
    X, Y, train_size=0.8, random_state=9)
```

```python
# 构建模型训练数据
gnb = GaussianNB()
gnb.fit(x_train, y_train)

# 查看结果评分
print('训练集准确率：', gnb.score(x_train, y_train))
print('测试机准确率：', gnb.score(x_test, y_test))
print('测试集准确率：', accuracy_score(gnb.predict(x_test), y_test))
```

```python
# 使用多项式朴素贝叶斯
mnb = MultinomialNB(alpha=0.01)
mnb.fit(X, Y)

# 查看结果评分
print('训练集准确率：', mnb.score(X, Y))
# print('测试机准确率：', mnb.score(x_test, y_test))
# print('测试集准确率：', accuracy_score(mnb.predict(x_test), y_test))
```

```python
# 伯努利分类
bnb = BernoulliNB()
bnb.fit(x_train, y_train)

# 查看结果评分
print('训练集准确率：', bnb.score(x_train, y_train))
print('测试机准确率：', bnb.score(x_test, y_test))
print('测试集准确率：', accuracy_score(bnb.predict(x_test), y_test))
```
