```python
print(__doc__)
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets, mixture  # 依次导入聚类、数据集、混合建模模块
from sklearn.neighbors import kneighbors_graph  # KNN算法
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
```

```python
plt.rcParams['font.sans-serif'] = ['Heiti']
plt.rcParams['axes.unicode_minus'] = False
```

#### 生成模拟数据集，生成足够大的数据集，但也不能太大，避免长时间运行

```python
# 随机数种子
np.random.seed(9)
```

```python
N = 1500    # 样本总量

# 生成圆环数据，
noisy_circles = datasets.make_circles(n_samples=N, factor=0.5, noise=0.05)
# 生成月牙数据
noisy_moons= datasets.make_moons(n_samples=N, noise=0.05)
blobs = datasets.make_blobs(n_samples=N, n_features=2, random_state=9)
no_strcuture = np.random.rand(N, 2), None     # 元祖

# 各项异性分布数据
X, Y = datasets.make_blobs(n_samples=N, random_state=170)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation) 
aniso = X_aniso, Y

# 生成具有不同方差的数据点
varied = datasets.make_blobs(n_samples=N, cluster_std=[1.0, 2.5, 0.5],
                             random_state=170)
```

#### 查看原始数据

```python
# 查看原始数据‘
plt.figure(facecolor='w', figsize=(20, 4))
plt.subplot(1, 5, 1)
plt.scatter(noisy_circles[0][:, 0], noisy_circles[0][:, 1], 
            c=noisy_circles[1], cmap='cool')
plt.title('圆环数据集')

plt.subplot(1, 5, 2)
plt.scatter(noisy_moons[0][:, 0], noisy_moons[0][:, 1], 
            c=noisy_moons[1], cmap='cool')
plt.title('月牙数据集')

plt.subplot(1, 5, 3)
plt.scatter(blobs[0][:, 0], blobs[0][:, 1], c=blobs[1], cmap='cool')
plt.title('blobs 数据集')

plt.subplot(1, 5, 4)
plt.scatter(aniso[0][:, 0], aniso[0][:, 1], c=aniso[1], cmap='cool')
plt.title('aniso 数据集')

plt.subplot(1, 5, 5)
plt.scatter(varied[0][:, 0], varied[0][:, 1], c=varied[1], cmap='cool')
plt.title('不同方差的数据集')

plt.show()
```

### 设置集群参数

```python
# 画布配置
plt.figure(figsize=(9 * 2 + 3, 12.5), facecolor='w')
plt.subplots_adjust(left=.02, right=.98, bottom=.001, top=.96, wspace=.05,
                    hspace=.01)

plot_num = 1
# 参数设置,初始参数
default_base = {
    'quantile': .3,
    'eps': .3,
    'damping': .9,
    'preference': -200,
    'n_neighbors': 10,
    'n_clusters': 3,
    'min_samples': 20,
    'xi': 0.05,
    'min_cluster_size': 0.1
}

# 不同数据集对应的不同参数的调整
datasets = [
    (noisy_circles, {'damping': .77, 'preference': -240,
                     'quantile': .2, 'n_clusters': 2,
                     'min_samples': 20, 'xi': 0.25}),
    (noisy_moons, {'damping': .75, 'preference': -220, 'n_clusters': 2}),
    (varied, {'eps': .18, 'n_neighbors': 2,
              'min_samples': 5, 'xi': 0.035, 'min_cluster_size': .2}),
    (aniso, {'eps': .15, 'n_neighbors': 2,
             'min_samples': 20, 'xi': 0.1, 'min_cluster_size': .2}),
    (blobs, {}),
    (no_strcuture, {})
]

for i_dataset, (dataset, algo_params) in enumerate(datasets):
    params = default_base.copy()
    params.update(algo_params)    # 将初始参数更新
    
    X, Y = dataset
    # 数据标准化
    X = StandardScaler().fit_transform(X)
    
    # 创建 bandwidth means
    bandwidth = cluster.estimate_bandwidth(X, quantile=params['quantile'])
    
    # 计算距离
    connectivity = kneighbors_graph(X, n_neighbors=params['n_neighbors'], 
                                   include_self=False)
    connectivity = 0.5 * (connectivity + connectivity.T)
    
    # 创建聚类模型
    ms = cluster.MeanShift(bandwidth=bandwidth, bin_seeding=True)
    two_means = cluster.MiniBatchKMeans(n_clusters=params['n_clusters'])
    # 凝聚的层次聚类算法
    ward = cluster.AgglomerativeClustering(n_clusters=params['n_clusters'],
                                           linkage='ward',
                                           connectivity=connectivity)
    # spectral
    spectral = cluster.SpectralClustering(
        n_clusters=params['n_clusters'], eigen_solver='arpack',
        affinity='nearest_neighbors')
    
    dbscan = cluster.DBSCAN(eps=params['eps'])
    optics = cluster.OPTICS(min_samples=params['min_samples'],
                           xi=params['xi'],
                           min_cluster_size=params['min_cluster_size'])
    affinity_propagation = cluster.AffinityPropagation(
        damping=params['damping'], preference=params['preference'])
    
    average_linkage = cluster.AgglomerativeClustering(
        linkage="average", affinity="cityblock",
        n_clusters=params['n_clusters'], connectivity=connectivity)
    
    birch = cluster.Birch(n_clusters=params['n_clusters'])
    
    gmm = mixture.GaussianMixture(
        n_components=params['n_clusters'], covariance_type='full')
    
    clustering_algorithms = (
        ('MiniBatchKMeans', two_means),
        ('AffinityPropagation', affinity_propagation),
        ('MeanShift', ms),
        ('SpectralClustering', spectral),
        ('Ward', ward),
        ('AgglomerativeClustering', average_linkage),
        ('DBSCAN', dbscan),
        ('OPTICS', optics),
        ('Birch', birch),
        ('GaussianMixture', gmm)
    )
    
    for name, algorithm in clustering_algorithms:
        t0 = time.time()

        # catch warnings related to kneighbors_graph
        with warnings.catch_warnings():
            warnings.filterwarnings(
                "ignore",
                message="the number of connected components of the " +
                "connectivity matrix is [0-9]{1,2}" +
                " > 1. Completing it to avoid stopping the tree early.",
                category=UserWarning)
            warnings.filterwarnings(
                "ignore",
                message="Graph is not fully connected, spectral embedding" +
                " may not work as expected.",
                category=UserWarning)
            algorithm.fit(X)

        t1 = time.time()
        if hasattr(algorithm, 'labels_'):
            y_pred = algorithm.labels_.astype(np.int)
        else:
            y_pred = algorithm.predict(X)

        plt.subplot(len(datasets), len(clustering_algorithms), plot_num)
        if i_dataset == 0:
            plt.title(name, size=18)

        colors = np.array(list(islice(cycle(['#377eb8', '#ff7f00', '#4daf4a',
                                             '#f781bf', '#a65628', '#984ea3',
                                             '#999999', '#e41a1c', '#dede00']),
                                      int(max(y_pred) + 1))))
        # add black color for outliers (if any)
        colors = np.append(colors, ["#000000"])
        plt.scatter(X[:, 0], X[:, 1], s=10, color=colors[y_pred])

        plt.xlim(-2.5, 2.5)
        plt.ylim(-2.5, 2.5)
        plt.xticks(())
        plt.yticks(())
        plt.text(.99, .01, ('%.2fs' % (t1 - t0)).lstrip('0'),
                 transform=plt.gca().transAxes, size=15,
                 horizontalalignment='right')
        plot_num += 1

plt.show()
```
