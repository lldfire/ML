```python
import time
import numpy as np
from itertools import cycle
import matplotlib.pyplot as plt
import matplotlib.colors as colors     # matplotlib 中的颜色

from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import Birch
from sklearn.preprocessing import StandardScaler
```

```python
plt.rcParams['font.sans-serif'] = ['Heiti']
plt.rcParams['axes.unicode_minus'] = False
```

### cluster.Birch()
#### 作用：
构造一个树数据结构，从叶子结点上读取簇的簇中心

#### 参数说明
- threshold：float,0.5；簇半径的阈值，达到该阈值后开始构造新的簇，阈值取值越小越能促进分裂
- branching_factor：int,50；节点中最大的集群数量，
- n_clusters: int,3；簇的数量
- compute_labels: bool,True，为每个簇计算拟合的标签
- copy: 是否覆盖原始数据，

#### 属性
- root_: CF tree 的根节点
- dummy_leaf_: 指向叶子节点的指针
- subcluster_centers_: 从叶子节点中读取簇的中心
- subcluster_labesl_: 聚类完成后的簇的标签
- labels_: 数据标签

#### 方法


### 比较 Birch 和 MiniBatchMeans 算法

#### 创建模拟数据

```python
# 模拟数据
xx = np.linspace(-22, 22, 10)    # 等差数列
yy = np.linspace(-22, 22, 10)

xx, yy = np.meshgrid(xx, yy)    # 矩阵扩展，扩展结果与矩阵大小有关
# 延水平方向将多个数组堆叠起来，组成一个二维矩阵
n_centers = np.hstack((np.ravel(xx)[:, np.newaxis],
                       np.ravel(yy)[:, np.newaxis]))


# 产生10万条特征为2，簇数为100的样本
X, Y = make_blobs(n_samples=100000, n_features=2,
                  centers=n_centers, random_state=9)
```

```python
# 构建模型
birch = Birch(threshold=0.7, n_clusters=100, branching_factor=100)
birch.fit(X)
```

#### 查看原始数据

```python
labels = birch.labels_
cent = birch.subcluster_centers_
fig = plt.figure(facecolor='w', figsize=(12, 6))
fig.subplots_adjust(left=0.04, right=0.98, bottom=0.1, top=0.9)


plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='winter')
plt.title('原始数据', fontsize=18, color='m')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='winter')
plt.title('聚类结果', fontsize=18, color='m')

plt.show()
```

#### 绘图比较

```python
# 使用matplotlib 中的所有颜色
colors_ = cycle(colors.cnames.keys())
fig = plt.figure(facecolor='w', figsize=(12, 4))
fig.subplots_adjust(left=0.04, right=0.98, bottom=0.1, top=0.9)

# 分别使用和不使用聚类数量来进行Birch训练模型
birch_models = [
    Birch(threshold=1.7, n_clusters=None),
    Birch(threshold=1.7, n_clusters=100)
]
final_step = ['不使用簇集', '使用簇集']

for idx, (birch_model, info) in enumerate(zip(birch_models, final_step)):
    start_time = time.time()
    birch_model.fit(X)
    time_ = time.time() - start_time
    print('Birch %s模型耗时%.2f秒' % (info, time_))
    
    # 绘制结果
    labels = birch_model.labels_     # 模型标签
    centroids = birch_model.subcluster_centers_     # 模型中心
    n_clusters = np.unique(labels).size      # 簇的数量
    print("n_clusters : %d" % n_clusters)
    
    # 设置子图图号
    ax = fig.add_subplot(1, 3, idx + 1)
    for this_centroid, k, col in zip(centroids, range(n_clusters), colors_):
        mask = labels == k    # 取出labels中等于k的值
        ax.scatter(X[mask, 0], X[mask, 1],
                   c='w', edgecolor=col, marker='.', alpha=0.5)
        if birch_model.n_clusters is None:
            ax.scatter(this_centroid[0], this_centroid[1], marker='+',
                       c='k', s=25)

    ax.set_xlim([-25, 25])
    ax.set_ylim([-25, 25])
    ax.set_title(info)
    
# 使用 MinibatchMeans 计算
start_time0 = time.time()
mbk = MiniBatchKMeans(init='k-means++', n_clusters=100, batch_size=100, 
                     n_init=10, max_no_improvement=10, verbose=False,
                      random_state=9)
mbk.fit(X)
print(f'MiniBatchMeans 算法耗时{round(( time.time() - start_time0), 2)}秒')
mbk_labels = np.unique(mbk.labels_).size    # 簇心的大小

ax = fig.add_subplot(1, 3, 3)
for this_centroid, k, col in zip(mbk.cluster_centers_, range(mbk_labels),
                                 colors_):
    mask = mbk.labels_ == k
    ax.scatter(X[mask, 0], X[mask, 1],
              c='w', edgecolor=col, marker='.', alpha=0.5)
    ax.scatter(this_centroid[0], this_centroid[1], c='k', marker='+', s=25)
    
ax.set_xlim([-25, 25])
ax.set_ylim([-25, 25])
ax.set_title(u'MiniBatchMeans 算法')

plt.show()
```
