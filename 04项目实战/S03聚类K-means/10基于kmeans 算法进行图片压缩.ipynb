```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from mpl_toolkits.mplot3d import Axes3D
from scipy import misc
from time import time

from sklearn.cluster import KMeans
# 计算点与点之间的最小距离
from sklearn.metrics import pairwise_distances_argmin 
from sklearn.datasets import load_sample_image
from sklearn.utils import shuffle
```

### pairwise_distances_argmin
#### 作用
计算一个点与一组点之间的最小距离

#### 参数说明
- X: 包含点的数组
- Y: 包含点的数组
- axis: 默认 1
- metric: 用于计算距离的矩阵


```python
# 图片是有3维的像素点组成的，使用聚类算法找到这些同样一类的像素点，以簇心的坐标代表这一类像素点
plt.rcParams['font.sans-serif'] = ['Heiti']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 读取图片
image = Image.open('./datas/xiaoren.png')
plt.figure(facecolor='w')
# plt.axis('off')
plt.imshow(image)
plt.show()
```

```python
# 加载图片将图片转为RGB格式，
xiaoren = np.array(image, np.float64) / 255
print(xiaoren.shape)

# 获取图片的长宽像素
original_shape = tuple(xiaoren.shape)
w, h, d = original_shape

# 将图片设置为 2维的
image_v = np.reshape(xiaoren, (w*h, d))
# 打乱样本,截取前10000行
image_sample = shuffle(image_v, random_state=9)[:10000]
```

```python
# 构建算法
t0 = time()
km = KMeans(n_clusters=128, init='k-means++', random_state=9)
km.fit(image_sample)
print(f'模型训练时间{time() - t0:.3f}秒')
```

```python
# 进行算法预测，预测所有图片数据的类别信息
t0 = time()
labels = km.predict(image_v)
print(f"算法预测时间为{time()-t0:.2f}秒")
```

```python
# 随机抽取数据，得到图片对象的聚类信息‘
code_randome = shuffle(image_v, random_state=9)[:128]
# 计算所有点和 这一组随机点之间的最小距离
labels_random = pairwise_distances_argmin(image_v, code_randome)
```

```python
# 重构一个压缩图片
def recreate_image(codebook, labels, w, h):
    """
    codebook:各个类别像素点的集合中心
    labels: 聚类的标签值
    w: 图片的宽度
    h: 图片的高度
    """
    d = codebook.shape[1]
    image = np.zeros((w, h, d))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            # 获取原始像素点对应的类别中心点的坐标
            # 再根据中心坐标获取像素值
            image[i][j] = codebook[labels[label_idx]]
            label_idx += 1
    return image
```

```python
# 图片的显示
# 原始图片
plt.figure(facecolor='w')
plt.clf()
# plt.axis('off')

if d == 1:
    plt.imshow(xiaoren, cmap=plt.cm.gray)
else:
    plt.imshow(xiaoren)
    
plt.savefig('datas/result_1.png')

# kmeans 算法压缩的图片
plt.figure(facecolor='w')
plt.clf()
# # plt.axis('off')
km_image = recreate_image(km.cluster_centers_, labels, w, h)
# km_image.shape = original_shape
if d == 1:
    plt.imshow(km_image, cmap=plt.cm.gray)
else:
    plt.imshow(km_image)
plt.savefig('datas/km_iamge.png')
    
# 随机抽取像素点压缩图片
plt.figure(facecolor='w')
plt.clf()
random_image = recreate_image(code_randome, labels_random, w, h)
random_image.shape = original_shape

if d == 1:
    plt.imshow(random_image, cmap=plt.cm.gray)
else:
    plt.imshow(random_image)
plt.savefig('datas/random_image.png')

# misc.imsave('datas/result_1.png', image)
# misc.imsave('datas/result_2.png', km_image)
# misc.imsave('datas/result_3.png', random_image)
```
